```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go

##Show plots in Colab
import plotly.io as pio
pio.renderers.default = "colab"
```

```python
def m_normalize(df):
  adf= (df-df.min())/(df.max()-df.min())
  return adf
```

#Power Consumption Dataset

```python
df_main=pd.read_csv("/content/drive/MyDrive/BDA 696 Project/Datasets/processsed/YMH_Clean_UCI_Power_Consumption_Dataset.csv")
df_main.head()
```

##Grouping Data by Date

```python
df_main['Date'] = pd.to_datetime(df_main['Date'])
df_main_2009=df_main.query("Date>='2009-01-01' and Date<='2009-12-31'")
df_main_2009=df_main_2009[['Date','Global_active_power'	,'Global_reactive_power'	,'Voltage'	,'Global_intensity'	,'Sub_metering_1'	,'Sub_metering_2'	,'Sub_metering_3']]
df_main_2009['Global_active_power']   = df_main_2009['Global_active_power'].apply(pd.to_numeric, errors = 'coerce').dropna().astype(float)
df_main_2009['Global_reactive_power'] = df_main_2009['Global_reactive_power'].apply(pd.to_numeric, errors = 'coerce').dropna().astype(float)
df_main_2009['Voltage']               = df_main_2009['Voltage'].apply(pd.to_numeric, errors = 'coerce').dropna().astype(float)
df_main_2009['Global_intensity']      = df_main_2009['Global_intensity'].apply(pd.to_numeric, errors = 'coerce').dropna().astype(float)
df_main_2009['Sub_metering_1']        = df_main_2009['Sub_metering_1'].apply(pd.to_numeric, errors = 'coerce').dropna().astype(float)
df_main_2009['Sub_metering_2']        = df_main_2009['Sub_metering_2'].apply(pd.to_numeric, errors = 'coerce').dropna().astype(float)
df_main_2009['Sub_metering_3']        = df_main_2009['Sub_metering_3'].apply(pd.to_numeric, errors = 'coerce').dropna().astype(float)
df_main_2009_sum=df_main_2009.groupby(pd.Grouper(key='Date', axis=1, freq='D'),as_index=True).sum().reset_index()
df_main_2009_sum.head()
```

```python
df_main_2009_sum.query("Date>='2009-06' and Date<'2009-07'").head()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = df_main_2009_sum['Date'],y=df_main_2009_sum['Global_reactive_power'], name = 'Actual'))
fig.update_layout(xaxis_title = 'Date', yaxis_title = 'Global_reactive_power')
fig.show()
```

#Month Based Overlay Chart for all Datasets

```python
df_gt1_sum=df_gt1.groupby(pd.Grouper(key='Date', axis=1, freq='M'),as_index=True).sum().reset_index()
df_gt1_sum['Hits']=m_normalize(df_gt1_sum['Hits'])

df_gt2_sum=df_gt2.groupby(pd.Grouper(key='Date', axis=1, freq='M'),as_index=True).sum().reset_index()
df_gt2_sum['Hits']=m_normalize(df_gt2_sum['Hits'])

df_gt3_sum=df_gt3.groupby(pd.Grouper(key='Date', axis=1, freq='M'),as_index=True).sum().reset_index()
df_gt3_sum['Hits']=m_normalize(df_gt3_sum['Hits'])

df_cpi_sum=df_cpi.groupby(pd.Grouper(key='Date', axis=1, freq='M'),as_index=True).sum().reset_index()
df_cpi_sum['CPI']=m_normalize(df_cpi_sum['CPI'])

df_cci_sum=df_cci.groupby(pd.Grouper(key='Date', axis=1, freq='M'),as_index=True).sum().reset_index()
df_cci_sum['CCI']=m_normalize(df_cci_sum['CCI'])


fig = go.Figure()
fig.add_trace(go.Scatter(x = df_main_2009_sum['Date'],y=df_main_2009_sum['Global_reactive_power'], mode = 'lines', name = 'Consumption'))
fig.add_trace(go.Scatter(x = df_gt1_sum['Date'],y=df_gt1_sum['Hits'], mode = 'lines', name = 'Google Trend 1'))
fig.add_trace(go.Scatter(x = df_gt2_sum['Date'],y=df_gt2_sum['Hits'], mode = 'lines', name = 'Google Trend 2'))
fig.add_trace(go.Scatter(x = df_gt3_sum['Date'],y=df_gt3_sum['Hits'], mode = 'lines', name = 'Google Trend 3'))
fig.add_trace(go.Scatter(x = df_cpi_sum['Date'],y=df_cpi_sum['CPI'], mode = 'lines', name = 'CPI'))
fig.add_trace(go.Scatter(x = df_cci_sum['Date'],y=df_cci_sum['CCI'], mode = 'lines', name = 'CCI'))

fig.update_layout(xaxis_title = 'Date', yaxis_title = 'Hits')
fig.show()
```
